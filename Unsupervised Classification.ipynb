```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#A glance to the dataset
df=pd.read_csv(r"CC GENERAL.csv")
df
```

```python
#check for NaN variables
df.isna().sum()
```

```python
#Eliminate the rows with NaN
df=df.dropna()
df
```

```python
#Customers ID is not relevant
df=df.iloc[:,1:]
df
```

```python
#Visualization of the variables and correlations
sns.pairplot(df)
```

```python
#standardize the data to normal distribution that is needed for both PCA and Clustering
from sklearn import preprocessing
df_normal = preprocessing.scale(df)
df_normal = pd.DataFrame(df_normal)
```

```python
#PCA will be used to visualize the data
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df_normal)
```

```python
#Varince explained by the first two components (2d graph --> first two PCs)
pca.explained_variance_ratio_
```

```python
# Which variable in the first component is more important
pd.Series(data=np.absolute(pca.components_[0]),index=df.columns,name="pca1").idxmax()
```

```python
# Which variable in the second component is more important
pd.Series(data=np.absolute(pca.components_[1]),index=df.columns,name="pca1").idxmax()
```

```python
# First Principal Component
pca_1=pca.fit_transform(df_normal)[:,0]
pca_1
```

```python
# Second Principal Component
pca_2=pca.fit_transform(df_normal)[:,1]
pca_2
```

```python
# First try k-means and determine the number of clusters from the "elbow"
plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_normal)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset with number of cluster enqual to 4
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(df_normal)
# New Dataframe called cluster
cluster = pd.DataFrame(y_kmeans)
# Adding cluster to the Dataset1
df_1=df
df_1['cluster'] = pd.DataFrame(data=cluster.values,index=df_1.index)
#Mean of clusters
kmeans_mean_cluster = pd.DataFrame(round(df_1.groupby('cluster').mean(),1))
kmeans_mean_cluster
```

```python
#Plot with two variables that better explain the variance (from pca analysis)
plt.figure(figsize=(10, 8))
plt.scatter(df['PURCHASES'], df['CASH_ADVANCE'],c=pd.Series(np.transpose(cluster.values)[0]), cmap='jet')  # plot points with cluster dependent colors
plt.title('Bank CC')
plt.show()
```

```python
#Use the first two PCs to have visually rapresent the data
plt.figure(figsize=(10, 8))
plt.scatter(pca_1, pca_2,c=pd.Series(np.transpose(cluster.values)[0]), cmap='jet')  # plot points with cluster dependent colors
plt.title('Bank CC')
plt.show()
```

```python
# Hierarchical clustering for the same dataset
from scipy.cluster.hierarchy import dendrogram, linkage
%matplotlib inline
#creating the linkage matrix
H_cluster = linkage(df_normal,'ward')
plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
dendrogram(
    H_cluster,
    truncate_mode='lastp', 
    p=20,  # show only the last p merged clusters
    leaf_rotation=90.,
    leaf_font_size=12.,
    show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.show()
```

```python
# Assigning the clusters and plotting the observations as per hierarchical clustering with k = 4
from scipy.cluster.hierarchy import fcluster
k=4
cluster_2 = fcluster(H_cluster, k, criterion='maxclust')
plt.figure(figsize=(10, 8))
plt.scatter(df['PURCHASES'], df['CASH_ADVANCE'],c=cluster_2, cmap='gist_rainbow')  # plot points with cluster dependent colors
plt.title('Bank CC')
plt.show()
```

```python
#Use the first two PCs to have visually rapresent the data
plt.figure(figsize=(10, 8))
plt.scatter(pca_1, pca_2,c=cluster_2, cmap='gist_rainbow')  # plot points with cluster dependent colors
plt.title('Bank CC')
plt.show()
```

```python
# Adding the hierarchical clustering to dataset
cluster_Hierarchical = pd.DataFrame(cluster_2)
df_2=df
df_2['cluster'] = pd.DataFrame(data=cluster_Hierarchical.values,index=df_2.index)
df_2.head()
```

```python
#Mean of clusters for hierarchical clustering
hierarchical_mean_cluster = pd.DataFrame(round(df_2.groupby('cluster').mean(),1))
hierarchical_mean_cluster
```

```python
# Kmean clustering is more precise into subdividing the differen groups.
# Now it is possible to proceed and characterize the different types of customers
kmeans_mean_cluster
```

```python
# Check the distribution graphycally
for c in df_1.iloc[:,:-1]:
    g=sns.FacetGrid(df_1, col='cluster')
    g.map(plt.hist,c)
```

# Customer Types

* Cluster 0: **Saving customers**: high balance and low purchase frequency, there is a low cash flow in the account. This customers utilized these funds to give cash in advance.
* Cluster 1: **Low-usage customers**: the is some balance in the account but is not used if not to have cash in advance.
* Cluster 2: **Wealthy customers**: high balance, high purchase amounth and high purchase frequency, there is an high cash flow in the account.
* Cluster 3: **Bank account customers**: this type of customers that use the credit card frequently, especially for installements.
